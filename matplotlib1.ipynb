```python
# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

# A1: Matplotlib is a popular data visualization library in Python used to create a wide range of static, interactive, and animated plots. It provides an interface similar to that of MATLAB, making it easy for users to create various types of plots.

# Matplotlib is used to visualize data in a graphical format, making it easier to interpret and understand patterns, trends, and relationships within the data. It is extensively used in data analysis, scientific research, machine learning, and other fields.

# Five plots that can be plotted using the Pyplot module of Matplotlib are:

# Line Plot: Used to display data points connected by straight lines, suitable for showing trends over time or continuous data.
# Scatter Plot: Used to display individual data points as dots, helpful in visualizing the relationship between two continuous variables.
# Bar Plot: Used to compare categorical data by representing each category with rectangular bars of varying heights.
# Histogram: Used to visualize the distribution of a continuous variable by dividing the data into intervals(bins) and showing the frequency of values in each bin.
# Pie Chart: Used to represent parts of a whole, where each category's proportion is represented as a slice of a circular pie.
```

```python
# A scatter plot is a type of plot that displays individual data points as dots on a two-dimensional plane.
# It is often used to visualize the relationship between two continuous variables. 
# Each dot represents a data point, with the x-coordinate indicating the value of one variable 
# and the y-coordinate representing the value of the other variable.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plotting the scatter plot
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating the subplots
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, plot 1
plt.plot(x, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)  # 2 rows, 2 columns, plot 2
plt.plot(x, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)  # 2 rows, 2 columns, plot 3
plt.plot(x, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)  # 2 rows, 2 columns, plot 4
plt.plot(x, y4)
plt.title("Line 4")

plt.tight_layout()  # Adjusts the spacing between subplots
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.figure(figsize=(8, 5))  # Adjust the figure size (optional)
plt.bar(company, profit)
plt.title("Company Profits")
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")
plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 5))  # Adjust the figure size (optional)
plt.barh(company, profit)
plt.title("Company Profits")
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating the box plot
plt.boxplot([box1, box2])
plt.title("Box Plot")
plt.xlabel("Box")
plt.ylabel("Value")
plt.show()
```
