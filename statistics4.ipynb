```python
# 1

import scipy.stats as stats

sample_mean = 50
sample_std = 5
confidence_level = 0.95
sample_size = 100

margin_of_error = stats.t.ppf(
    (1 + confidence_level) / 2, sample_size - 1) * (sample_std / (sample_size ** 0.5))
confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)

confidence_interval
```

```python
#2
import scipy.stats as stats

observed = [15, 25, 20, 60, 30]
# Multiplying by total observed to get expected counts
expected = [0.2, 0.2, 0.2, 0.1, 0.1, 0.2] * sum(observed)

chi2, p_value = stats.chisquare(observed, expected)

chi2, p_value
```

```python
#4
import statsmodels.stats.proportion as proportion

sample_size = 500
smokers = 60

conf_interval = proportion.proportion_confint(
    smokers, sample_size, alpha=0.05, method='normal')

conf_interval
```

```python
#5
import scipy.stats as stats

sample_mean = 75
sample_std = 12
sample_size = 100

confidence_interval = stats.norm.interval(
    0.90, loc=sample_mean, scale=sample_std/(sample_size ** 0.5))

confidence_interval
```

```python
#6
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

df = 10

x = np.linspace(0, 30, 500)
y = stats.chi2.pdf(x, df)

plt.plot(x, y)
plt.fill_between(x[x >= 15], y[x >= 15], 0, alpha=0.3, color='blue')
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability density')
plt.title('Chi-square Distribution (df=10)')
plt.show()
```

```python
#8
import scipy.stats as stats

observed_tails = 45
expected_tails = 50  # Expected tails for a fair coin

observed = [observed_tails, 100 - observed_tails]
expected = [expected_tails, 100 - expected_tails]

chi2, p_value = stats.chisquare(observed, expected)

chi2, p_value
```
