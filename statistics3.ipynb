```python
# Q1: Estimation statistics is a branch of statistics that involves making inferences or predictions 
# about population parameters based on sample data. It allows us to estimate unknown population parameters,
# such as the mean or proportion, using information from a sample.

# Point estimate: A point estimate is a single numerical value that is used to estimate the population 
# parameter of interest. It is usually based on sample data and provides an estimate of the unknown population parameter. For example, the sample mean can be used as a point estimate of the population mean.

# Interval estimate: An interval estimate provides a range of values within which the population parameter
# is likely to lie. It takes into account the uncertainty associated with estimating the parameter based 
# on limited sample data. The most common form of interval estimate is the confidence interval. 
# It consists of an upper and lower bound, calculated based on the sample data and a specified level of 
# confidence. The confidence level represents the percentage of intervals that would contain the true
# population parameter if the sampling process was repeated many times.


```

```python
# Q2
import math


def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    standard_error = sample_std_dev / math.sqrt(sample_size)
    # Assuming a 95% confidence level (standard normal distribution)
    z_critical = 1.96
    margin_of_error = z_critical * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    return lower_bound, upper_bound
```

```python
# Q3: Hypothesis testing is a statistical technique used to make inferences about a population
# based on a sample of data. It involves formulating two competing hypotheses, the null hypothesis(H0) 
# and the alternative hypothesis(H1 or Ha), and examining the evidence in the sample to determine which
# hypothesis is more likely.

# The null hypothesis(H0) represents a statement of no effect, no difference, or no relationship in the
# population. The alternative hypothesis(H1 or Ha) represents the contrary to the null hypothesis, 
# suggesting that there is an effect, difference, or relationship in the population.

# Hypothesis testing is used to assess the strength of evidence against the null hypothesis and make 
# decisions based on that evidence. It helps determine if observed differences or relationships in the 
# sample are statistically significant and likely to exist in the population.

# The importance of hypothesis testing lies in its ability to provide a systematic and rigorous approach 
# to decision-making based on data. It allows researchers and analysts to draw conclusions and make 
# inferences about populations, helping to guide scientific research, inform policy decisions, and support
# evidence-based decision-making.
```

```python
# Q4: Hypothesis:
# The average weight of male college students is greater than the average weight of female college students.

# Null hypothesis(H0): The average weight of male college students is not greater than the average weight of female college students.
# Alternative hypothesis(H1): The average weight of male college students is greater than the average weight of female college students.
```

```python
# Q 5
import numpy as np
from scipy.stats import ttest_ind


def hypothesis_test_two_sample(sample1, sample2, alpha):
    t_statistic, p_value = ttest_ind(sample1, sample2)
    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")


# Example usage
sample1 = [2, 4, 6, 8, 10]
sample2 = [1, 3, 5, 7, 9]
alpha = 0.05

hypothesis_test_two_sample(sample1, sample2, alpha)
```

```python
# Q6: Null hypothesis(H0): It represents a statement of no effect, no difference, or no relationship in the population. It assumes that any observed differences or relationships in the sample are due to random chance or sampling variability.

# Alternative hypothesis(H1 or Ha): It represents the contrary to the null hypothesis. It suggests that there is an effect, difference, or relationship in the population. It states the possibility of a non-random pattern or a specific effect.

# Examples:

# Null hypothesis: The mean height of male adults is equal to the mean height of female adults.
# Alternative hypothesis: The mean height of male adults is different from the mean height of female adults.

# Null hypothesis: There is no difference in test scores between two groups of students.
# Alternative hypothesis: There is a difference in test scores between two groups of students.

# Null hypothesis: The new treatment has no effect on patient recovery.
# Alternative hypothesis: The new treatment improves patient recovery.

# Q7: The steps involved in hypothesis testing are as follows:

# State the null hypothesis(H0) and the alternative hypothesis(H1 or Ha).
# Choose the significance level(alpha) to determine the threshold for rejecting the null hypothesis.
# Collect and analyze the sample data.
# Calculate the test statistic appropriate for the hypothesis being tested(e.g., t-statistic, z-statistic).
# Determine the p-value associated with the test statistic. The p-value represents the probability of obtaining the observed sample data, assuming the null hypothesis is true.
# Compare the p-value to the significance level(alpha).
# If the p-value is less than alpha, reject the null hypothesis and conclude that there is evidence to support the alternative hypothesis.
# If the p-value is greater than or equal to alpha, fail to reject the null hypothesis and conclude that there is not enough evidence to support the alternative hypothesis.
# Interpret the results in the context of the problem and draw conclusions.
# Q8: The p-value is a probability value that measures the strength of evidence against the null hypothesis. It represents the probability of obtaining the observed sample data, or more extreme data, assuming the null hypothesis is true. The p-value helps determine the level of statistical significance of the results.

# In hypothesis testing, the significance level (alpha) is predetermined, typically set to 0.05 (5 % ) or 0.01 (1%). The p-value is then compared to the significance level to make a decision:

# If the p-value is less than alpha(p < alpha), it suggests that the observed data is unlikely to occur by chance alone, assuming the null hypothesis is true. In this case, the null hypothesis is rejected, and the alternative hypothesis is supported. The result is considered statistically significant.
# If the p-value is greater than or equal to alpha(p >= alpha), it suggests that the observed data is reasonably likely to occur by chance alone, assuming the null hypothesis is true.In this case, the null hypothesis is not rejected, and there is not enough evidence to support the alternative hypothesis. The result is not considered statistically significant.
# The significance of the p-value lies in its ability to quantify the strength of evidence against the null hypothesis. A smaller p-value indicates stronger evidence against the null hypothesis and suggests that the observed results are unlikely to be due to random chance. On the other hand, a larger p-value indicates weaker evidence against the null hypothesis and suggests that the observed results could plausibly occur by random chance.
```

```python
import numpy as np
from scipy.stats import ttest_ind


def two_sample_t_test(sample1, sample2, alpha):
    t_statistic, p_value = ttest_ind(sample1, sample2)
    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")


# Example usage
sample1 = [2, 4, 6, 8, 10]
sample2 = [1, 3, 5, 7, 9]
alpha = 0.05

two_sample_t_test(sample1, sample2, alpha)
```

```python
# Q11: The Student's t-distribution, or t-distribution for short, is a probability distribution that is used when the sample size is small or when the population standard deviation is unknown. It is similar to the standard normal distribution(Z-distribution), but it has thicker tails, allowing for greater variability in small samples.

# The t-distribution is used in situations where the population standard deviation is not known and must be estimated from the sample data. It is commonly used for hypothesis testing and constructing confidence intervals when working with small samples.

# The t-distribution is characterized by its degrees of freedom, which depend on the sample size. As the sample size increases, the t-distribution approaches the standard normal distribution. When the sample size is large(typically above 30), the t-distribution and the standard normal distribution are practically equivalent.

# Q12: The t-statistic is a measure of how much the sample mean deviates from the hypothesized population mean, in terms of the standard error. It is used in hypothesis testing to determine whether the difference between the sample mean and the hypothesized population mean is statistically significant.

# The formula for calculating the t-statistic for a sample mean is:
# t = (sample_mean - hypothesized_mean) / \
#     (sample_standard_deviation / sqrt(sample_size))

# Where:

# sample_mean is the mean of the sample
# hypothesized_mean is the hypothesized population mean(under the null hypothesis)
# sample_standard_deviation is the standard deviation of the sample
# sample_size is the size of the sample
# The t-statistic follows a t-distribution with (sample_size - 1) degrees of freedom when the null hypothesis is true.
```

```python
import scipy.stats as stats

sample_mean = 500
sample_std_dev = 50
sample_size = 50
confidence_level = 0.95

# Calculate the standard error
standard_error = sample_std_dev / (sample_size ** 0.5)

# Calculate the margin of error
margin_of_error = stats.t.ppf(
    (1 + confidence_level) / 2, sample_size - 1) * standard_error

# Calculate the lower and upper bounds of the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print the confidence interval
print("Population mean revenue with a 95% confidence interval:")
print(f"Lower bound: ${lower_bound:.2f}")
print(f"Upper bound: ${upper_bound:.2f}")
```
