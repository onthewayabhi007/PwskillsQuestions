```python

# Answer 1:
# Here are five plots that can be created using the Seaborn library along with their uses:

# Scatter Plot:

# Uses: To display the relationship between two numerical variables.
# Seaborn function: scatterplot()
# Line Plot:

# Uses: To visualize the trend or changes in numerical data over continuous intervals.
# Seaborn function: lineplot()
# Bar Plot:

# Uses: To compare categorical data using rectangular bars.
# Seaborn function: barplot()
# Histogram:

# Uses: To represent the distribution of a single numerical variable.
# Seaborn function: histplot()
# Box Plot:

# Uses: To display the distribution of a numerical variable across different categories.
# Seaborn function: boxplot()
```

```python
import seaborn as sns

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Plot the line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

```python
import seaborn as sns

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Plot the box plots
sns.boxplot(x="pclass", y="age", data=titanic)
sns.boxplot(x="pclass", y="fare", data=titanic)
```

```python
import seaborn as sns

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Plot the histogram
sns.histplot(data=diamonds, x='price', hue='cut')
```

```python
import seaborn as sns

# Load the iris dataset
iris = sns.load_dataset("iris")

# Plot the pair plot
sns.pairplot(data=iris, hue="species")
```
